## Introdução
Este relatório apresenta uma análise exploratória dos dados de clientes de uma empresa de telecomunicações com o objetivo de identificar padrões e fatores que contribuem para a evasão de clientes (Churn). O problema do Churn é crítico para qualquer negócio baseado em assinaturas, pois reter clientes existentes é geralmente mais custo-efetivo do que adquirir novos. Através da análise de diversas características dos clientes e de seus comportamentos, buscamos entender as causas da evasão e propor recomendações para mitigar esse problema.
## Limpeza e Tratamento de Dados
Os dados foram obtidos a partir de um arquivo JSON hospedado no GitHub.
Os passos de limpeza e tratamento incluíram:
- **Importação:** Os dados foram carregados utilizando a biblioteca pandas a partir da URL fornecida.
- **Normalização:** Como o arquivo JSON possuía uma estrutura aninhada, utilizamos pd.json_normalize para achatar a estrutura e criar um DataFrame tabular.
- **Tratamento de Valores Categóricos:** Valores como 'Yes'/'No' e 'Female'/'Male' em colunas relevantes foram substituídos por 1s e 0s para facilitar a análise numérica e a modelagem futura. Termos como 'internet.InternetService', 'account.Contract' e 'phone.MultipleLines' foram convertidos para minúsculas para padronização.
- **Conversão de Tipos de Dados:** As colunas relacionadas a cobranças ('account.Charges.Total', 'account.Charges.Monthly') foram convertidas para tipos numéricos. Valores que não puderam ser convertidos foram tratados (erros foram convertidos para NaN e removidos).
- **Criação de Novas Features:** Uma coluna 'Contas_diarias' foi criada para representar a média de cobrança diária com base na cobrança mensal. Medidas estatísticas (média, mediana, desvio padrão) da coluna 'Contas_diarias' foram calculadas e adicionadas ao DataFrame.

## Conclusões e Insights
Com base na análise exploratória, podemos extrair os seguintes insights principais:
- **Contratos Mensais:** Clientes em contratos mensais são o grupo de maior risco para evasão. A flexibilidade pode ser conveniente, mas não promove a lealdade a longo prazo.
- **Métodos de Pagamento:** O método de pagamento 'Electronic check' está associado a uma alta taxa de Churn (Desistência). Pode haver problemas de usabilidade, confiança ou um perfil de cliente diferente associado a este método.
- **Serviços Adicionais de Internet:** A ausência de serviços de segurança online e suporte técnico parece estar correlacionada com uma maior propensão à evasão. Isso sugere que esses serviços podem ser importantes para a satisfação e retenção do cliente.
- **Total Gasto:** Clientes com menor histórico de gastos totais tendem a evadir mais. Isso pode indicar que clientes com maior tempo de casa ou que utilizam mais serviços são mais retidos.


## Recomendações
Com base nos insights obtidos, sugiro as seguintes ações para reduzir a evasão de clientes:
- **Incentivar Contratos de Longo Prazo:** Oferecer descontos ou benefícios adicionais para clientes que optam por contratos de 1 ou 2 anos. Comunicar o valor de permanecer com a empresa a longo prazo.
- **Analisar o Método de Pagamento 'Electronic Check':** Investigar se há problemas específicos ou frustrações associadas ao método de pagamento 'Electronic check' que possam estar contribuindo para a evasão. Considerar oferecer alternativas ou melhorar a experiência com este método.
- **Promover Serviços de Segurança e Suporte:** Destacar o valor e os benefícios dos serviços de segurança online e suporte técnico. Oferecer pacotes atraentes que incluam esses serviços, especialmente para novos clientes com internet.
- **Programas de Fidelidade:** Desenvolver programas de fidelidade que recompensem clientes com maior tempo de casa ou que utilizam múltiplos serviços. Isso pode aumentar o custo de 'troca' para o cliente.
- **Monitoramento Proativo:** Utilizar os insights da análise para identificar clientes em risco de evasão e implementar campanhas de retenção proativas, como ofertas personalizadas ou contato direto.

### Relevância das Variáveis na Árvore de Decisão

As Árvores de Decisão fornecem uma métrica de "importância da variável" que indica o quanto cada variável contribui para a redução da impureza (como Gini impurity) nos nós da árvore. Variáveis com valores mais altos são consideradas mais importantes para a classificação.

Abaixo, visualizamos a importância das variáveis calculada pela Árvore de Decisão treinada com `max_depth=3`:

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importances = pd.Series(arvore.feature_importances_, index=one_hot.get_feature_names_out(colunas))
sorted_feature_importances = feature_importances.sort_values(ascending=False)


top_6_features = sorted_feature_importances.head(6)

plt.figure(figsize=(12, 8))

sns.barplot(x=top_6_features.values, y=top_6_features.index)
plt.title('Top 6 Variáveis Mais Relevantes na Árvore de Decisão (max_depth=3)')
plt.xlabel('Importância')
plt.ylabel('Variável')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('barplot_variaveisimportantes_arvore.png')
plt.show()
```

Do gráfico da importância das variáveis na Árvore de Decisão, podemos observar claramente quais variáveis tiveram o maior impacto na divisão dos nós e, consequentemente, na tomada de decisão do modelo para classificar um cliente como "Churn" ou "Não Churn".

### Relevância das Variáveis no Modelo KNN

O modelo KNN (K-Nearest Neighbors) não calcula a importância explícita das variáveis da mesma forma que modelos baseados em árvores. No KNN, a relevância de uma variável está relacionada à sua contribuição para o cálculo da distância entre os pontos de dados. Variáveis que apresentam maior variação nos dados (após a normalização, se aplicada) e que são mais discriminativas entre as classes terão um impacto maior na determinação dos vizinhos mais próximos e, portanto, na classificação.

Considerando que o KNN ajustado utilizou a distância de Manhattan (`metric='manhattan'`) e que aplicamos a normalização (`MinMaxScaler`) aos dados, todas as variáveis no conjunto de dados transformado (`x_treino_normalizado`, `x_teste_normalizado`) contribuem para o cálculo da distância.

Com base nas análises exploratórias anteriores e na importância das variáveis na Árvore de Decisão, podemos inferir que as variáveis mais relevantes para o KNN são provavelmente aquelas que mostraram uma clara distinção entre clientes que evadiram e os que não evadiram nos gráficos iniciais.

# Relatório Detalhado: Fatores de Evasão de Clientes e Estratégias de Retenção

## Introdução

Este relatório consolida as descobertas da análise de dados e modelagem preditiva para identificar os principais fatores que levam à evasão de clientes (Churn) em uma empresa de telecomunicações. Com base nesses insights, propomos estratégias direcionadas para melhorar a retenção de clientes.

## Análise de Fatores de Evasão

Nossa análise exploratória e a avaliação dos modelos de classificação (Árvore de Decisão e KNN) destacaram diversas variáveis que influenciam significativamente a probabilidade de um cliente evadir.

### Fatores Relevantes Identificados:

Com base na importância das variáveis na Árvore de Decisão e na distinção observada nos gráficos de análise exploratória (relevantes para o KNN), os principais fatores que afetam a evasão são:

1.  **Tipo de Contrato (`account.Contract`):** Clientes com contratos de curto prazo, especialmente contratos **"month-to-month"**, apresentam uma taxa de evasão consideravelmente maior em comparação com clientes com contratos de um ou dois anos. Isso foi o fator mais importante na Árvore de Decisão.

2.  **Tempo de Permanência (`customer.tenure`):** Clientes com **menor tempo de permanência** na empresa são mais propensos a evadir. A distribuição do tempo de contrato mostrou que os clientes que evadiram geralmente tinham um tempo de permanência mais curto.

3.  **Serviço de Internet (`internet.InternetService`):** O tipo de serviço de internet, particularmente o **"fiber optic"**, parece estar associado a uma maior taxa de evasão, conforme indicado pela importância na Árvore de Decisão.

4.  **Cobrança Mensal (`account.Charges.Monthly`):** A cobrança mensal também se mostrou relevante na Árvore de Decisão, sugerindo que clientes com **cobranças mensais mais altas** podem ter uma propensão maior à evasão, possivelmente devido a insatisfação com o custo-benefício.

5.  **Total Gasto (`account.Charges.Total`):** Clientes com **menor histórico de gastos totais** ao longo do tempo são mais propensos a evadir. Isso está alinhado com a observação sobre o tempo de permanência, pois clientes com maior tempo de casa tendem a ter um total gasto maior.

6.  **Serviços Adicionais de Internet (`internet.OnlineSecurity`, etc.):** A ausência de certos serviços de segurança online ou suporte técnico também apareceu como um fator relevante, indicando que a percepção de segurança e suporte pode influenciar a decisão de permanecer.

## Desempenho dos Modelos

A avaliação dos modelos confirmou a dificuldade em prever a evasão devido ao desequilíbrio da classe.

*   O **Dummy Classifier** (linha de base) teve uma acurácia de ~73.4%, simplesmente prevendo a classe majoritária.
*   A **Árvore de Decisão (max_depth=3)** obteve uma acurácia de ~78.9%, com uma precisão razoável para a classe churn ~69.2% mas recall mais baixo ~36.9%.
*   O **KNN Inicial** teve acurácia de ~75.5% e um F1-score para churn de ~52.7%, mostrando algum overfitting.
*   O **Melhor KNN ** melhorou o desempenho do KNN inicial, alcançando uma acurácia de ~77.4% e um F1-score para churn de ~56.0%, com recall maior ~54.2% em comparação com a Árvore de Decisão.

O KNN Ajustado se mostrou o modelo mais promissor para identificar clientes que evadirão (melhor F1-score e Recall), enquanto a Árvore de Decisão foi mais precisa quando previu churn.

## Estratégias de Retenção Propostas

Com base nos fatores identificados, as seguintes estratégias de retenção são recomendadas:

1.  **Foco em Clientes com Contratos Mensais:**
    *   Oferecer incentivos atraentes (descontos, benefícios extras) para clientes em contratos mensais migrarem para planos de 1 ou 2 anos.
    *   Comunicar proativamente o valor e a economia de longo prazo de contratos mais longos.

2.  **Programas de Boas-Vindas e Engajamento Inicial:**
    *   Implementar programas de acompanhamento intensivo para novos clientes, especialmente aqueles com menor tempo de permanência, para garantir que estejam satisfeitos com os serviços e resolver quaisquer problemas iniciais.

3.  **Análise e Melhoria do Serviço de Fibra Óptica:**
    *   Investigar as causas da maior evasão entre clientes de fibra óptica. Pode estar relacionado a problemas de instalação, qualidade do serviço percebida, suporte técnico ou expectativas não atendidas.
    *   Realizar pesquisas de satisfação específicas com este grupo de clientes.

4.  **Avaliação da Estrutura de Cobrança:**
    *   Analisar se a estrutura de cobrança é clara e competitiva, especialmente para planos com cobranças mensais mais altas.
    *   Considerar oferecer pacotes ou promoções que melhorem o custo-benefício percebido.

5.  **Promoção de Serviços Adicionais:**
    *   Destacar os benefícios dos serviços de segurança online e suporte técnico.
    *   Considerar incluir esses serviços em pacotes promocionais ou oferecer testes gratuitos para aumentar a adesão.

6.  **Monitoramento Preditivo e Ação Proativa:**
    *   Utilizar o modelo de classificação (preferencialmente o KNN Ajustado devido ao seu melhor Recall) para identificar clientes com alta probabilidade de evasão.
    *   Desenvolver campanhas de retenção personalizadas para esses clientes, como ofertas especiais, contato direto para resolver problemas ou pesquisas de satisfação direcionadas.

A evasão de clientes é um desafio significativo, mas a análise dos dados revelou padrões claros e fatores influentes. Ao focar nos clientes de maior risco (contratos mensais, menor tempo de permanência, usuários de fibra óptica, etc.) e implementar estratégias de retenção direcionadas com base nesses insights, a empresa pode reduzir a taxa de Churn e melhorar a saúde geral da base de clientes.
